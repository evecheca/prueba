# [dbo].[sp_consultar_carreras]

```sql
/****** Object:  StoredProcedure [dbo].[sp_consultar_carreras]    Script Date: 11/9/2022 10:50:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create proc [dbo].[sp_consultar_carreras]
as
begin
	select id_carrera, nombre from carreras
	order by id_carrera
end
GO
```

# [dbo].[sp_consultar_detalleCarreras]

```sql
/****** Object:  StoredProcedure [dbo].[sp_consultar_detalleCarreras]    Script Date: 11/9/2022 10:50:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure  [dbo].[sp_consultar_detalleCarreras]
as
begin
	select c.id_carrera 'codigo',c.nombre as 'Carrera',c.titulo 'titulo',
	a.id_asignatura 'codigo asignatura',
	a.nombre  'Nombre Asignatura',
	dc.anio_cursado as 'Año Cursado', 
	dc.cuatrimestre as 'Cuatrimestre'
	from detalleCarreras as dc 
	inner join asignaturas as a on dc.id_asignatura = a.id_asignatura
	inner join carreras as c on dc.id_carrera = c.id_carrera
	WHERE c.eliminado=1
	order by c.id_carrera
end
GO
```

# [dbo].[sp_insertar_carrera]

```sql
/****** Object:  StoredProcedure [dbo].[sp_insertar_carrera]    Script Date: 11/9/2022 10:50:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_insertar_carrera]
@nombre varchar(100),
@titulo varchar(100),
@new_id_carrera int output
as
begin
	insert into carreras(nombre,titulo) values(@nombre,@titulo );
	set @new_id_carrera = SCOPE_IDENTITY();
end
GO
```

# [dbo].[sp_insertar_detalleCarreras]

```sql
/****** Object:  StoredProcedure [dbo].[sp_insertar_detalleCarreras]    Script Date: 11/9/2022 10:50:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create proc [dbo].[sp_insertar_detalleCarreras]
@anioCursado int,
@cuatrimestre int,
@id_carrera int,
@id_asignatura int

as
begin
	insert into DetalleCarreras(anio_cursado, cuatrimestre, id_carrera, id_asignatura)
	values(@anioCursado, @cuatrimestre, @id_carrera, @id_asignatura)
end
GO
```

# [dbo].[SP_MostrarCarrera]

```sql
/****** Object:  StoredProcedure [dbo].[SP_MostrarCarrera]    Script Date: 11/9/2022 10:50:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[SP_MostrarCarrera]
 
 As
 begin select *from Carrera 
 where eliminado=1
 end
GO
```

# [dbo].[sp_registrar_baja_carrera]

```sql
/****** Object:  StoredProcedure [dbo].[sp_registrar_baja_carrera]    Script Date: 11/9/2022 10:50:36 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_registrar_baja_carrera]
 @id_carrera int
 As
 begin
 Update Carreras 
 SET eliminado=1
 Where id_Carrera= @id_carrera
 end
GO
```
